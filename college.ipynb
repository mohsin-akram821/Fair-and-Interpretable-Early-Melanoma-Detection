```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

# Example: directory with 6 images
img_dir = "College"  

# List of 6 images (replace with your filenames)
img_files = [
    "College\Screenshot 2025-09-11 094301.png",
    "College\Screenshot 2025-09-11 094322.png",
    "College\Screenshot 2025-09-11 094343.png",
    "College\Screenshot 2025-09-11 094430.png",
    "College\Screenshot 2025-09-11 094809.png",
    "College\Screenshot 2025-09-11 094846.png"
]

# Corresponding titles
titles = [
    "Benign", "Benign", "Benign",
    "Benign", "Malignant", "Malignant"
]

# Create 2x3 grid
fig, axes = plt.subplots(2, 3, figsize=(10, 6))

for i, ax in enumerate(axes.flat):
    img_path = os.path.join(img_dir, img_files[i])
    img = mpimg.imread(img_path)
    ax.imshow(img)
    ax.set_title(titles[i], fontsize=12)
    ax.axis("off")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import os

# Base directory (only once)
img_dir = r"College"

# Just filenames here
img_files = [
    "Screenshot 2025-09-11 094301.png",
    "Screenshot 2025-09-11 094322.png",
    "Screenshot 2025-09-11 094343.png",
    "Screenshot 2025-09-11 094430.png",
    "Screenshot 2025-09-11 094809.png",
    "Screenshot 2025-09-11 094846.png"
]

titles = ["Malignant", "Benign", "Benign", "Malignant", "Malignant", "Benign"]

fig, axes = plt.subplots(2, 3, figsize=(8, 6))

# Choose a fixed axis size (e.g., 0300 in both directions)
xlim, ylim = (0, 250), (0, 250)

for i, ax in enumerate(axes.flat):
    img_path = os.path.join(img_dir, img_files[i])
    img = mpimg.imread(img_path)

    ax.imshow(img, extent=[0, 250, 0, 250])  # scale into same axis
    ax.set_title(titles[i], fontsize=12)
    
    # Set same axis limits for all
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
    
    # Keep box but remove tick labels
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout()
plt.show()
```
