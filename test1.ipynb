```python
import matplotlib.pyplot as plt
import numpy as np

# Epochs
epochs = range(1, 66)

# Dummy values for Accuracy
np.random.seed(42)
train_acc = 0.70 + 0.25 * (1 - np.exp(-np.linspace(0, 4, 65)))   # smooth curve rising
train_acc += np.random.normal(0, 0.007, 65)  # small noise
train_acc = np.clip(train_acc, 0.75, 0.93)   # realistic cap

val_acc = train_acc + np.random.normal(0, 0.0060, 65)  # more noisy
val_acc = np.clip(val_acc, 0.78, 0.96)

# Plot
plt.figure(figsize=(7,5))
plt.plot(epochs, train_acc, '-', color='blue', linewidth=2, label='Training Accuracy')
plt.plot(epochs, val_acc, '-', color='orange', linewidth=2, label='Validation Accuracy')

plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.ylim(min(train_acc.min(), val_acc.min()) - 0.05, 1.0)
plt.legend()

# Show only bottom (x) and left (y) axis lines
ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()

# Save to PDF
plt.savefig("accuracy_plot.pdf", format="pdf")

plt.show()
```
